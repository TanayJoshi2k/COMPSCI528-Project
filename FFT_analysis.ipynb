```python
import numpy as np
import pandas as pd
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
```

```python
def fast_fourier(data, normalize=''):
    # Extract individual sensor data (gyro and accelerometer)
    gyro_x = data['gyro_x'].values
    gyro_y = data['gyro_y'].values
    gyro_z = data['gyro_z'].values
    acce_x = data['acce_x'].values
    acce_y = data['acce_y'].values
    acce_z = data['acce_z'].values

    # Number of samples (N) and sampling interval (T)
    N = len(gyro_x)  # Assuming all columns have the same number of samples
    T = 1/100  # Replace with the actual sampling rate (T = 1 / sampling frequency)

    # Apply FFT on each axis (gyro_x, gyro_y, gyro_z, acce_x, acce_y, acce_z)
    def perform_fft(signal, N, T):
        # FFT of the signal
        yf = fft(signal)
        xf = fftfreq(N, T)[:N//2]  # Frequency bins
        fft_values = 2.0/N * np.abs(yf[0:N//2])  # Scaled FFT values
        
        # Normalization based on the chosen method
        if normalize == 'zscore':
            mean = np.mean(fft_values)
            std = np.std(fft_values)
            fft_values = (fft_values - mean) / std  # Z-score normalization
        elif normalize == 'minmax':
            min_val = np.min(fft_values)
            max_val = np.max(fft_values)
            fft_values = (fft_values - min_val) / (max_val - min_val)  # Min-Max normalization
        else:
            xf = xf
            fft_values = fft_values
        
        return xf, fft_values

    # Perform FFT for each sensor axis
    xf_gyro_x, fft_gyro_x = perform_fft(gyro_x, N, T)
    xf_gyro_y, fft_gyro_y = perform_fft(gyro_y, N, T)
    xf_gyro_z, fft_gyro_z = perform_fft(gyro_z, N, T)
    xf_acce_x, fft_acce_x = perform_fft(acce_x, N, T)
    xf_acce_y, fft_acce_y = perform_fft(acce_y, N, T)
    xf_acce_z, fft_acce_z = perform_fft(acce_z, N, T)

    # Plotting the FFT of each axis (Gyroscope and Accelerometer)
    plt.figure(figsize=(12, 8))

    # Plot Gyroscope data FFT
    plt.subplot(2, 1, 1)
    plt.plot(xf_gyro_x, fft_gyro_x, label='Gyro X')
    plt.plot(xf_gyro_y, fft_gyro_y, label='Gyro Y')
    plt.plot(xf_gyro_z, fft_gyro_z, label='Gyro Z')
    plt.title(f'FFT of Gyroscope Data (Normalization: {normalize.capitalize()})')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()

    # Plot Accelerometer data FFT
    plt.subplot(2, 1, 2)
    plt.plot(xf_acce_x, fft_acce_x, label='Acce X')
    plt.plot(xf_acce_y, fft_acce_y, label='Acce Y')
    plt.plot(xf_acce_z, fft_acce_z, label='Acce Z')
    plt.title(f'FFT of Accelerometer Data (Normalization: {normalize.capitalize()})')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()
```

```python
right = pd.read_csv("Data/UP_gesture_data_1.csv")
fast_fourier(right)
```
