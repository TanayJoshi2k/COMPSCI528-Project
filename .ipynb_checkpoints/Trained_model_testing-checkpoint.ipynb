```python
def spectrogram_and_plot(data):
    # Extract individual sensor data (gyro and accelerometer)
    gyro_x = data['gyro_x'].values
    gyro_y = data['gyro_y'].values
    gyro_z = data['gyro_z'].values
    acce_x = data['acce_x'].values
    acce_y = data['acce_y'].values
    acce_z = data['acce_z'].values

    # Number of samples (N) and sampling interval (T)
    N = len(gyro_x)  # Assuming all columns have the same number of samples
    T = 1/100  # Replace with the actual sampling rate (T = 1 / sampling frequency)
    fs = 1/T

    # Plotting the time-domain data for gyroscope and accelerometer
    plt.figure(figsize=(12, 20))

    # Plot Gyroscope time-domain data
    plt.subplot(8, 1, 1)
    plt.plot(np.arange(N) * T, gyro_x, label='Gyro X')
    plt.plot(np.arange(N) * T, gyro_y, label='Gyro Y')
    plt.plot(np.arange(N) * T, gyro_z, label='Gyro Z')
    plt.title('Time Domain - Gyroscope Data')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()

    # Plot Accelerometer time-domain data
    plt.subplot(8, 1, 2)
    plt.plot(np.arange(N) * T, acce_x, label='Acce X')
    plt.plot(np.arange(N) * T, acce_y, label='Acce Y')
    plt.plot(np.arange(N) * T, acce_z, label='Acce Z')
    plt.title('Time Domain - Accelerometer Data')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.grid()

    # Generate and plot spectrograms for the gyroscope data
    plt.subplot(8, 1, 3)
    f_gyro, t_gyro, Sxx_gyro = spectrogram(gyro_x,fs=fs,nperseg=40,noverlap=10)
    plt.pcolormesh(t_gyro, f_gyro, 10*np.log(Sxx_gyro), shading='gouraud')
    plt.colorbar(label='Power/Frequency (dB/Hz)')
    plt.title('Spectrogram - Gyroscope X-axis')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')

    # Generate and plot spectrograms for the accelerometer data (e.g., acce_x)
    plt.subplot(8, 1, 4)
    f_gyro, t_gyro, Sxx_gyro = spectrogram(gyro_y,fs=fs,nperseg=40,noverlap=10)
    plt.pcolormesh(t_gyro, f_gyro, 10*np.log(Sxx_gyro), shading='gouraud')
    plt.colorbar(label='Power/Frequency (dB/Hz)')
    plt.title('Spectrogram - Gyroscope Y-axis')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')
    
    plt.subplot(8, 1, 5)
    f_gyro, t_gyro, Sxx_gyro = spectrogram(gyro_z,fs=fs,nperseg=40,noverlap=10)
    plt.pcolormesh(t_gyro, f_gyro, 10*np.log(Sxx_gyro), shading='gouraud')
    plt.colorbar(label='Power/Frequency (dB/Hz)')
    plt.title('Spectrogram - Gyroscope Z-axis')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')
    
    plt.subplot(8, 1, 6)
    f_acce, t_acce, Sxx_acce = spectrogram(acce_x,fs=fs,nperseg=40,noverlap=10)
    plt.pcolormesh(t_gyro, f_gyro, 10*np.log(Sxx_acce), shading='gouraud')
    plt.colorbar(label='Power/Frequency (dB/Hz)')
    plt.title('Spectrogram - Accelerometer X-axis')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')

    # Generate and plot spectrograms for the accelerometer data (e.g., acce_x)
    plt.subplot(8, 1, 7)
    f_acce, t_acce, Sxx_acce = spectrogram(acce_y,fs=fs,nperseg=40,noverlap=10)
    plt.pcolormesh(t_gyro, f_gyro, 10*np.log(Sxx_acce), shading='gouraud')
    plt.colorbar(label='Power/Frequency (dB/Hz)')
    plt.title('Spectrogram - Accelerometer Y-axis')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')
    
    plt.subplot(8, 1, 8)
    f_acce, t_acce, Sxx_acce = spectrogram(acce_z,fs=fs,nperseg=40,noverlap=10)
    plt.pcolormesh(t_gyro, f_gyro, 10*np.log(Sxx_acce), shading='gouraud')
    plt.colorbar(label='Power/Frequency (dB/Hz)')
    plt.title('Spectrogram - Accelerometer Z-axis')
    plt.ylabel('Frequency (Hz)')
    plt.xlabel('Time (s)')

    plt.tight_layout()
    #plt.savefig("Images/"+filename.split("/")[1].split(".")[0]+".png", dpi=300)
    
    plt.show()
```

```python
def load_data(files,fixed_length):
    features,labels = [],[]
    csv = True
    for file in files:
        if ".csv" in file:
            label = file.split("_")[0].split("\\")[1]
            df = pd.read_csv(file)
        else:
            print(file)
            csv = False
            label = file.split("/")[2].split("\\")[0]
            with open(file, 'r') as file:
                lines = file.readlines()
            
            # Split lines into header and data
            header = lines[0].strip().split("\t")  # Assuming tab-separated values
            header = [head.strip() for head in header]
            data = [line.strip().split("\t") for line in lines[1:]]  # Skip header row

            # Convert data to a pandas DataFrame
            df = pd.DataFrame(data, columns=header)

        df[['gyro_x', 'gyro_y', 'gyro_z', 'acce_x', 'acce_y', 'acce_z']] = df[
            ['gyro_x', 'gyro_y', 'gyro_z', 'acce_x', 'acce_y', 'acce_z']
        ].apply(pd.to_numeric, errors='coerce')
        
        # Drop rows with any NaN values in the specified columns
        df = df.dropna(subset=['gyro_x', 'gyro_y', 'gyro_z', 'acce_x', 'acce_y', 'acce_z'])

        # Convert to numpy and normalize if there are any valid rows left
        if not df.empty:
            data = df[['gyro_x', 'gyro_y', 'gyro_z', 'acce_x', 'acce_y', 'acce_z']].values.astype(np.float32)
            data = (data - data.min(axis=0)) / (data.max(axis=0) - data.min(axis=0))  # Normalize data
            
            if len(data) > fixed_length:
                data = data[:fixed_length]  # truncate if longer than fixed length
            elif len(data) < fixed_length:
                padding = np.zeros((fixed_length - len(data), data.shape[1]), dtype=np.float32)
                data = np.vstack((data, padding)) 
                
            features.append(data.flatten())
        
    return np.array(features)
```
